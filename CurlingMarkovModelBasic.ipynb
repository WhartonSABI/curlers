```python
def determine_hammer(team_ids, end_ids, results, first_end_hammer):
    hammer_by_end = {1: first_end_hammer}
    current_hammer = first_end_hammer

    unique_ends = sorted(set(end_ids))
    last_end = unique_ends[-1]

    for i in range(0, len(results), 2):
        end = end_ids[i]

        t1, t2 = team_ids[i], team_ids[i + 1]
        r1, r2 = results[i], results[i + 1]

        # who scored
        if r1 > 0:
            scorer = t1
        elif r2 > 0:
            scorer = t2
        else:
            scorer = None

        # update hammer
        if scorer is not None:
            current_hammer = t2 if scorer == t1 else t1

        # only write hammer if next end exists
        if end < last_end:
            hammer_by_end[end + 1] = current_hammer

    return hammer_by_end
```

```python
def split_by_end(df, keep_all_ends=False):
    """
    Split curling game data so each row represents a single end.
    
    Parameters:
    df: DataFrame with game-level data where list columns contain per-end information
    keep_all_ends: If True, keeps all End1-9 columns in every row (for historical context)
                   If False, only includes data relevant to the specific end
    
    Returns:
    DataFrame with one row per end
    """
    rows = []
    
    for idx, row in df.iterrows():
        # Get the list columns - these contain per-end data (alternating teams)
        competition_ids = row['CompetitionID']
        session_ids = row['SessionID']
        game_ids = row['GameID']
        team_ids = row['TeamID']
        end_ids = row['EndID']
        results = row['Result']
        power_plays = row['PowerPlay']
        
        # The lists alternate between Team1 and Team2
        n_ends = len(end_ids) // 2
        
        # Create a row for each end
        for i in range(n_ends):
            team1_idx = i * 2
            team2_idx = i * 2 + 1
            
            end_num = end_ids[team1_idx]
            
            # Core data
            end_row = {
                'UniID': row['UniID'],
                'CompetitionID': competition_ids[team1_idx],
                'SessionID': session_ids[team1_idx],
                'GameID': game_ids[team1_idx],
                'EndID': end_num,
                'hammer_first_end': row['hammer_first_end'],
                'Team1': row['Team1'],
                'Team2': row['Team2'],
                'Team1_PowerPlay_End': row['Team1_PowerPlay_End'],
                'Team2_PowerPlay_End': row['Team2_PowerPlay_End'],
                'Team1_ID': team_ids[team1_idx],
                'Team1_Result': results[team1_idx],
                'Team1_PowerPlay': power_plays[team1_idx],
                'Team2_ID': team_ids[team2_idx],
                'Team2_Result': results[team2_idx],
                'Team2_PowerPlay': power_plays[team2_idx]
            }
            
            if keep_all_ends:
                # Keep ALL End1-9 columns (current behavior)
                for col in df.columns:
                    if col.startswith('End') and 'TeamWithHammer' in col:
                        end_row[col] = row[col]
                    elif col.startswith('End') and 'IsNotHammer' in col:
                        end_row[col] = row[col]
                    elif 'TotalPointsScoredByEnd' in col:
                        end_row[col] = row[col]
                    elif col.startswith('ScoreDiffByEnd'):
                        end_row[col] = row[col]
            else:
                # Only include data for THIS specific end
                if end_num <= 9:
                    end_row['TeamWithHammer'] = row[f'End{end_num}TeamWithHammer']
                    end_row['Team1_IsNotHammer'] = row[f'End{end_num}_Team1IsNotHammer']
                    end_row['Team2_IsNotHammer'] = row[f'End{end_num}_Team2IsNotHammer']
                    end_row['Team1_TotalPointsScoredByEnd'] = row[f'Team1TotalPointsScoredByEnd{end_num}']
                    end_row['Team2_TotalPointsScoredByEnd'] = row[f'Team2TotalPointsScoredByEnd{end_num}']
                    end_row['ScoreDiffByEnd'] = row[f'ScoreDiffByEnd{end_num}']
                    
                    # Also include previous end's score for context (optional)
                    if end_num > 1:
                        end_row['Team1_TotalPointsScoredByPrevEnd'] = row[f'Team1TotalPointsScoredByEnd{end_num-1}']
                        end_row['Team2_TotalPointsScoredByPrevEnd'] = row[f'Team2TotalPointsScoredByEnd{end_num-1}']
                        end_row['ScoreDiffByPrevEnd'] = row[f'ScoreDiffByEnd{end_num-1}']
            
            rows.append(end_row)
    
    return pd.DataFrame(rows)

# Usage with your ends1 dataframe:
# Option 1: Keep all End1-9 columns (your current output)
ends1_by_end_full = split_by_end(ends1, keep_all_ends=True)

# Option 2: Only relevant end data (cleaner)
ends1_by_end_clean = split_by_end(ends1, keep_all_ends=False)

print("Option 1 - Keep all End columns:")
print(ends1_by_end_full.head(10))
print(f"Shape: {ends1_by_end_full.shape}")

print("\n" + "="*80 + "\n")

print("Option 2 - Only relevant end data:")
print(ends1_by_end_clean.head(10))
print(f"Shape: {ends1_by_end_clean.shape}")
print(f"\nColumns: {ends1_by_end_clean.columns.tolist()}")
```

```python
import numpy as np
import pandas as pd

print("===== START CELL =====")

# ================================
# 1. START FROM CLEAN OPTION 2 DF
# ================================
df = ends1_by_end_clean.copy()
print("Loaded dataframe:", df.shape)

# Drop any previously broken score columns
df = df.drop(
    columns=["ScoreDiffByEnd", "ScoreDiffByPrevEnd"],
    errors="ignore"
)

# =====================================
# 2. REBUILD SCORE DIFFERENTIALS (TRUTH)
# =====================================
# Definition: score diff = Team2 âˆ’ Team1
df["ScoreDiffByEnd"] = (
    df["Team2_TotalPointsScoredByEnd"]
    - df["Team1_TotalPointsScoredByEnd"]
)

print("ScoreDiffByEnd rebuilt")

# ======================================
# 3. BUILD PREVIOUS-END SCORE (MARKOV)
# ======================================
df = df.sort_values(["UniID", "EndID"])

df["ScoreDiffByPrevEnd"] = (
    df.groupby("UniID")["ScoreDiffByEnd"]
      .shift(1)
)

# First end starts at 0 by definition
df.loc[df["EndID"] == 1, "ScoreDiffByPrevEnd"] = 0

print("ScoreDiffByPrevEnd rebuilt")

# =====================================
# 4. DEFINE MARKOV STATE VARIABLES
# =====================================
df["s"] = df["ScoreDiffByEnd"] - df["ScoreDiffByPrevEnd"]  # score change
df["e"] = 8 - df["EndID"]                                 # ends remaining
df["h"] = 1 - df["Team2_IsNotHammer"]                      # hammer indicator

print("Markov state variables created")

# =====================================
# 5. REMOVE POWER PLAY ENDS
# =====================================
df_model = df[
    df["Team1_PowerPlay"].isna() &
    df["Team2_PowerPlay"].isna()
].copy()

print("Filtered power play ends:", df_model.shape)

# =====================================
# 6. VALIDATION (MUST PASS)
# =====================================
assert np.allclose(
    df_model["ScoreDiffByPrevEnd"] + df_model["s"],
    df_model["ScoreDiffByEnd"],
    equal_nan=False
), "Score transition identity violated!"

print("Score transition identity validated")

# =====================================
# 7. BUILD TRANSITION KERNEL
# =====================================
transition_kernel = (
    df_model
    .groupby(["ScoreDiffByPrevEnd", "e", "h"])["s"]
    .value_counts(normalize=True)
)

print("Transition kernel built")
print("Kernel states:", len(transition_kernel))

transition_kernel.to_pickle("transition_kernel.pkl")
print("Transition kernel saved")

# =====================================
# 8. MONTE CARLO SIMULATOR
# =====================================
def simulate_game(x0, e0, h0, kernel, N=1000):  # TEMPORARILY LOWER N
    wins = 0
    for _ in range(N):
        x, e, h = x0, e0, h0
        while e > 0:
            try:
                dist = kernel.loc[(x, e, h)]
            except KeyError:
                break
            s = np.random.choice(dist.index, p=dist.values)
            x += s
            e -= 1
            h = 1 - h
        if x > 0:
            wins += 1
    return wins / N

print("Monte Carlo simulator ready")

# =====================================
# 9. COMPUTE WIN PROBABILITY STATES
# =====================================
states = []

for x in range(-8, 9):
    print(f"Simulating score diff x = {x}")
    for e in range(1, 9):
        for h in [0, 1]:
            wp = simulate_game(x, e, h, transition_kernel)
            states.append((x, e, h, wp))

print("All simulations completed")

wp_df = pd.DataFrame(
    states, columns=["x", "e", "h", "wp"]
)

print("Win probability dataframe created:", wp_df.shape)

print("===== END CELL =====")
```

```python

# ============================================================
# 1. WIN PROBABILITY LOOKUP TABLE
# ============================================================

wp_lookup = (
    wp_df
    .set_index(["x", "e", "h"])["wp"]
)

# ============================================================
# 2. WIN PROBABILITY IF PUNT (BLANK END)
# ============================================================

def wp_if_blank(x, e, h):
    # Terminal condition
    if e == 0:
        return 1.0 if x > 0 else 0.0
    return wp_lookup.get((x, e-1, 1-h), np.nan)

# ============================================================
# 3. WIN PROBABILITY IF FIGHT (EMPIRICAL TRANSITION)
# ============================================================

def wp_if_fight(x, e, h, kernel):
    if e == 0:
        return 1.0 if x > 0 else 0.0
    
    try:
        dist = kernel.loc[(x, e, h)]
    except KeyError:
        return np.nan

    wp = 0.0
    for s, p in dist.items():
        wp_next = wp_lookup.get((x + s, e-1, 1-h), np.nan)
        if not np.isnan(wp_next):
            wp += p * wp_next

    return wp

# ============================================================
# 4. COMPUTE OPTIMAL DECISION SURFACE
# ============================================================

decisions = []

for x in range(-8, 9):
    for e in range(1, 9):
        h = 1  # decision only when YOU have hammer
        
        wp_blank = wp_if_blank(x, e, h)
        wp_fight = wp_if_fight(x, e, h, transition_kernel)

        if np.isnan(wp_blank) or np.isnan(wp_fight):
            continue

        delta = wp_fight - wp_blank

        decisions.append({
            "x": x,
            "e": e,
            "wp_fight": wp_fight,
            "wp_blank": wp_blank,
            "delta": delta,
            "decision": "FIGHT" if delta > 0 else "PUNT"
        })

decision_df = pd.DataFrame(decisions)

# ============================================================
# 5. VISUALIZE DECISION BOUNDARY
# ============================================================

pivot = decision_df.pivot(
    index="e",
    columns="x",
    values="delta"
)

plt.figure(figsize=(10,6))
sns.heatmap(pivot, cmap="RdBu", center=0)
plt.title("Fight vs Punt Decision Boundary")
plt.xlabel("Score Differential (Team2 âˆ’ Team1)")
plt.ylabel("Ends Remaining")
plt.show()

# ============================================================
# 6. SANITY CHECKS (MUST BE TRUE)
# ============================================================

# 1. Fighting more attractive when behind
decision_df.groupby("x")["delta"].mean().plot(
    title="Average Fight Advantage vs Score Differential"
)
plt.show()

# 2. Fighting more attractive late
decision_df.groupby("e")["delta"].mean().plot(
    title="Average Fight Advantage vs Ends Remaining"
)
plt.show()

# ============================================================
# 7. EXAMPLE DECISIONS (FOR INTERPRETATION)
# ============================================================

print("Even score situations:")
print(decision_df.query("x == 0").sort_values("e"))

print("\nLeading by 2:")
print(decision_df.query("x == 2").sort_values("e"))

print("\nTrailing by 2:")
print(decision_df.query("x == -2").sort_values("e"))
```

```python
# =========================================
# COACH-FACING DECISION TABLE
# =========================================

# Bin score differential into interpretable categories
def score_bucket(x):
    if x <= 0:
        return "Tied or Trailing"
    elif x == 1:
        return "Leading by 1"
    elif x == 2:
        return "Leading by 2"
    else:
        return "Leading by 3+"

decision_df["score_bucket"] = decision_df["x"].apply(score_bucket)

# Bin ends remaining into early / late
decision_df["time_bucket"] = decision_df["e"].apply(
    lambda e: "Late (â‰¤2 ends)" if e <= 2 else "Early (â‰¥3 ends)"
)

# For each situation, choose the dominant decision
coach_table = (
    decision_df
    .groupby(["score_bucket", "time_bucket"])["delta"]
    .mean()
    .reset_index()
)

coach_table["Optimal Action"] = coach_table["delta"].apply(
    lambda d: "FIGHT" if d > 0 else "PUNT"
)

# Drop raw delta if you want a clean table
coach_table = coach_table.drop(columns="delta")

# Sort for readability
coach_table = coach_table.sort_values(
    ["score_bucket", "time_bucket"]
)

print(coach_table)
```

```python
# ============================================================
# STEP 4 â€” COST OF MAKING THE WRONG DECISION (REGRET ANALYSIS)
# ============================================================

import matplotlib.pyplot as plt
import seaborn as sns

# ------------------------------------------------------------
# 4.1 Compute regret (absolute win probability loss)
# ------------------------------------------------------------

decision_df["regret"] = decision_df["delta"].abs()

print("Top 10 most costly decision states:")
print(
    decision_df
    .sort_values("regret", ascending=False)
    [["x", "e", "decision", "wp_fight", "wp_blank", "regret"]]
    .head(10)
)

# ------------------------------------------------------------
# 4.2 Average regret by score differential
# ------------------------------------------------------------

plt.figure(figsize=(8,4))
decision_df.groupby("x")["regret"].mean().plot(marker="o")
plt.title("Average Cost of Wrong Decision vs Score Differential")
plt.xlabel("Score Differential (Team2 âˆ’ Team1)")
plt.ylabel("Win Probability Lost")
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 4.3 Average regret by ends remaining
# ------------------------------------------------------------

plt.figure(figsize=(8,4))
decision_df.groupby("e")["regret"].mean().plot(marker="o")
plt.title("Average Cost of Wrong Decision vs Ends Remaining")
plt.xlabel("Ends Remaining")
plt.ylabel("Win Probability Lost")
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 4.4 Heatmap of regret (where strategy matters most)
# ------------------------------------------------------------

regret_pivot = decision_df.pivot(
    index="e",
    columns="x",
    values="regret"
)

plt.figure(figsize=(10,6))
sns.heatmap(regret_pivot, cmap="Reds")
plt.title("Cost of Making the Wrong Decision (Win Probability Lost)")
plt.xlabel("Score Differential (Team2 âˆ’ Team1)")
plt.ylabel("Ends Remaining")
plt.show()

# ------------------------------------------------------------
# 4.5 Coach-facing high-impact decision table
# ------------------------------------------------------------

# Define high-impact as â‰¥5% win probability loss
decision_df["High_Impact"] = decision_df["regret"] >= 0.05

high_impact_table = (
    decision_df
    .query("High_Impact")
    .sort_values("regret", ascending=False)
    [["x", "e", "decision", "regret"]]
)

print("\nHigh-impact decision states (â‰¥5% WP swing):")
print(high_impact_table.head(15))
```
